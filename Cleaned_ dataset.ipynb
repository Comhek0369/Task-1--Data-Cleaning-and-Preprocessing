```python
import pandas as pd
```

```python
data = pd.read_csv("netflix_dataset.csv")
```

```python
data.head(10)
```

```python
data.columns
```

```python
data = data.drop(columns = "Cast")
```

```python
data.columns
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.head(10)
```

```python
data["Show_id"]= data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data
```

```python
data["Show_id"][0]
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```
